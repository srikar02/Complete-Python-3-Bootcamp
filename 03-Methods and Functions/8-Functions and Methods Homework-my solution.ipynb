**Write a function that checks whether a number is in a given range (inclusive of high and low)**

```python
def ran_check(num,low,high):
    if (num >= low and num <= high):
        return '{num} is in the range between {low} and {high}'.format(num=num,low=low,high=high)
    return '{num} is not in the range between {low} and {high}'.format(num=num,low=low,high=high)
```

```python
# Check
ran_check(5,2,7)
```

```python
ran_check(6,10,20)
```

```python
ran_check(21,7,21)
```

```python
ran_bool(3,1,10)
```

```python
ran_bool(6,1,5)
```

**Write a Python function that accepts a string and calculates the number of upper case letters and lower case letters.**

    Sample String : 'Hello Mr. Rogers, how are you this fine Tuesday?'
    Expected Output : 
    No. of Upper case characters : 4
    No. of Lower case Characters : 33

HINT: Two string methods that might prove useful: **.isupper()** and **.islower()**

If you feel ambitious, explore the Collections module to solve this problem!

```python
def up_low(s):
    upper_case_count = 0
    lower_case_count = 0
    for letter in s:
        if(letter.isupper()):
            upper_case_count += 1
        elif(letter.islower()):
            lower_case_count += 1
        else:
            pass
    print('Upper Case Characters: {}'.format(upper_case_count))
    print('Lower Case Characters: {}'.format(lower_case_count))
```

```python
s = 'Hello Mr. Rogers, how are you this fine Tuesday?'
up_low(s)
```

```python
s = 'Mary and Samantha arrived at the bus station early but waited until noon for the bus.'
up_low(s)
```

**Write a Python function to multiply all the numbers in a list.**

    Sample List : [1, 2, 3, -4]
    Expected Output : -24

```python
def multiply(numbers):
    total = 1
    for num in numbers:
        total *= num
    return total
```

```python
multiply([1,2,3,-4])
```

```python
multiply([1,2,3,-4,-6,11,-42,32,54])
```

**Write a Python function that checks whether a word or phrase is palindrome or not.**

Note: A palindrome is word, phrase, or sequence that reads the same backward as forward, e.g., madam,kayak,racecar, or a phrase "nurses run". Hint: You may want to check out the .replace() method in a string to help out with dealing with spaces. Also google search how to reverse a string in Python, there are some clever ways to do it with slicing notation.

```python
def palindrome(s):
    s = s.replace(' ','')
    reverse = s[::-1]
    return s == reverse
```

```python
palindrome('helleh')
```

```python
palindrome('nurses run')
```

**Write a Python function to check whether a string is pangram or not. (Assume the string passed in does not have any punctuation)**

    Note : Pangrams are words or sentences containing every letter of the alphabet at least once.
    For example : "The quick brown fox jumps over the lazy dog"

Hint: You may want to use .replace() method to get rid of spaces.

Hint: Look at the [string module](https://stackoverflow.com/questions/16060899/alphabet-range-in-python)

Hint: In case you want to use [set comparisons](https://medium.com/better-programming/a-visual-guide-to-set-comparisons-in-python-6ab7edb9ec41)

```python
import string

def ispangram(str1, alphabet=string.ascii_lowercase):
    str1 = str1.lower().replace(' ','')
    set_str1 = set(str1)
    set_alphabet = set(alphabet)
    diff = set_alphabet.difference(set_str1)
    print(diff)
    if (len(diff) == 0):
        print ('{} is a panagram'.format(str1))
    else:
        print ('{} is not a panagram'.format(str1))
```
